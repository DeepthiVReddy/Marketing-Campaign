```python
import pandas as pd
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Step 1: Load the dataset
file_path = "CuteOutfitMarketing.csv"  # Replace with your file path
data = pd.read_csv(file_path)

# Step 2: Ensure categorical variables are properly defined
categorical_columns = ['Age', 'Gender', 'OwnHome', 'Married', 'Location', 'History']
for col in categorical_columns:
    data[col] = data[col].astype('category')

# Step 3: Fit the logistic regression model
formula = "Purchase ~ Age + Gender + OwnHome + Married + Location + Children + History + Income"
logistic_model = smf.logit(formula=formula, data=data).fit()

# Display model summary
print(logistic_model.summary())

# Step 4: Add predicted probabilities to the dataset
data['PredictedProb'] = logistic_model.predict(data)

# Step 5: Create reversed deciles (1 = highest predicted probabilities, 10 = lowest)
data['Decile'] = 9 - pd.qcut(data['PredictedProb'], 10, labels=False, duplicates='drop')

# Step 6: Calculate cumulative metrics
positive_ratio = data['Purchase'].mean()
total_positive_responses = data['Purchase'].sum()

# Group data by decile and calculate cumulative values
decile_summary = data.groupby('Decile').agg(
    TotalResponses=('Purchase', 'sum'),
    TotalCustomers=('Purchase', 'count')
).sort_index(ascending=True).reset_index()

decile_summary['CumulativeResponses'] = decile_summary['TotalResponses'].cumsum()
decile_summary['CumulativeSamples'] = decile_summary['TotalCustomers'].cumsum()

# Calculate cumulative lift and cumulative gain
decile_summary['CumulativeLift'] = (
    decile_summary['CumulativeResponses'] /
    (positive_ratio * decile_summary['CumulativeSamples'])
)

decile_summary['CumulativeGain'] = (
    decile_summary['CumulativeResponses'] / total_positive_responses
)

# Step 7: Display the decile summary
print("Cumulative Metrics Summary:")
print(decile_summary)

# Step 8: Plot cumulative lift chart
plt.figure(figsize=(10, 6))
plt.plot(decile_summary['Decile'], decile_summary['CumulativeLift'], marker='o', label='Cumulative Lift', linestyle='-')
plt.title('Cumulative Lift Chart')
plt.xlabel('Decile (1 = Highest Predicted Probabilities)')
plt.ylabel('Cumulative Lift')
plt.grid(True)
plt.legend()
plt.show()

# Step 9: Plot cumulative gain chart
plt.figure(figsize=(10, 6))
plt.plot(decile_summary['Decile'], decile_summary['CumulativeGain'], marker='o', label='Cumulative Gain', linestyle='-')
plt.title('Cumulative Gain Chart')
plt.xlabel('Decile (1 = Highest Predicted Probabilities)')
plt.ylabel('Cumulative Gain')
plt.grid(True)
plt.legend()
plt.show()

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import roc_auc_score
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('CuteOutfitMarketing.csv')

# Separate the features and the target variable
X = data.drop('Purchase', axis=1)
y = data['Purchase']

# Identify categorical and numerical columns
categorical_cols = X.select_dtypes(include=['object', 'category']).columns
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns

# Preprocessing for numerical data
numeric_transformer = StandardScaler()

# Preprocessing for categorical data
categorical_transformer = OneHotEncoder(handle_unknown='ignore')

# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_cols),
        ('cat', categorical_transformer, categorical_cols)
    ])

# Define the model
model = GradientBoostingClassifier(
    learning_rate=0.1,
    n_estimators=100,
    max_depth=4,
    random_state=1
)

# Create and evaluate the pipeline
pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('model', model)
                          ])

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Fit the model
pipeline.fit(X_train, y_train)

# Predict the probabilities
probabilities = pipeline.predict_proba(X_test)[:, 1]

# Create deciles
data_test = pd.DataFrame({'Purchase': y_test, 'Prob': probabilities})
data_test['Decile'] = pd.qcut(data_test['Prob'], 10, labels=False)

# Calculate total responses and response rates for each decile
decile_summary = data_test.groupby('Decile').agg(
    total_responses=('Purchase', 'sum'),
    total=('Purchase', 'size')
)
decile_summary['response_rate'] = decile_summary['total_responses'] / decile_summary['total']

# Sort the deciles
decile_summary = decile_summary.sort_index(ascending=False)

print(decile_summary)

# Generate a decile lift chart
plt.plot(decile_summary.index + 1, decile_summary['response_rate'], marker='o')
plt.title('Decile Lift Chart')
plt.xlabel('Decile')
plt.ylabel('Response Rate')
plt.grid(True)
plt.show()
```


The Gradient Boosted Trees model was built using Purchase as the dependent variable and all other variables as independent variables. The model was trained using the specified parameters: learning rate = 0.1, number of estimators = 100, max depth = 4, and random seed = 1. Preprocessing was applied to categorical variables (using one-hot encoding) and numeric variables (using standard scaling). Probabilities were predicted on the test set, deciles were created, and the total responses and response rates for each decile were calculated.

<br>

The decile lift chart created using these results shows that higher deciles (9, 8, 7) have significantly higher response rates, indicating that the model performs well in identifying customers with a higher likelihood of purchase. However, lower deciles (4 and below) have a response rate of 0%, indicating that the model struggles to differentiate non-buyers effectively in those groups.

  
 <br>

The results suggest that the model is effective in identifying potential buyers in the top deciles, as Decile 9 has the highest response rate of 33.33%, followed by Decile 8 (23.33%) and Decile 7 (16.67%). This pattern demonstrates the predictive power of the Gradient Boosted Trees model in targeting high-probability buyers.


```python
# Calculate cumulative lift and gain
decile_summary['cumulative_responses'] = decile_summary['total_responses'].cumsum()
decile_summary['cumulative_total'] = decile_summary['total'].cumsum()
decile_summary['cumulative_response_rate'] = decile_summary['cumulative_responses'] / decile_summary['cumulative_total']

# Total number of responses and records
total_responses = decile_summary['total_responses'].sum()
total_records = decile_summary['total'].sum()

# Calculate gain and lift
decile_summary['gain'] = decile_summary['cumulative_responses'] / total_responses
decile_summary['lift'] = decile_summary['cumulative_response_rate'] / (total_responses / total_records)

print(decile_summary[['cumulative_responses', 'cumulative_response_rate', 'gain', 'lift']])

# Generate cumulative lift and gain charts
fig, ax1 = plt.subplots()

ax1.set_xlabel('Decile')
ax1.set_ylabel('Cumulative Gain', color='tab:blue')
ax1.plot(decile_summary.index + 1, decile_summary['gain'], marker='o', color='tab:blue', label='Cumulative Gain')
ax1.tick_params(axis='y', labelcolor='tab:blue')
ax1.set_xticks(range(1, 11))

ax2 = ax1.twinx()
ax2.set_ylabel('Cumulative Lift', color='tab:orange')
ax2.plot(decile_summary.index + 1, decile_summary['lift'], marker='x', color='tab:orange', label='Cumulative Lift')
ax2.tick_params(axis='y', labelcolor='tab:orange')

fig.tight_layout()
plt.title('Cumulative Gain and Lift Chart')
plt.xlabel('Decile')
plt.grid(True)
plt.show()
```

```python
# Assuming you have the cost per mail and profit per response
cost_per_mail = 2  # Replace with the actual cost
profit_per_response = 20  # Replace with the actual profit

# Calculate the breakeven response rate
breakeven_response_rate = cost_per_mail / profit_per_response
print(f'Breakeven Response Rate: {breakeven_response_rate:.4f}')

# Create the 'Mail2' variable
data_test['Mail2'] = (data_test['Prob'] >= breakeven_response_rate).astype(int)

# Calculate total mailed and total responses for 'Mail2' = 1
total_mailed = data_test['Mail2'].sum()
total_responses = data_test.loc[data_test['Mail2'] == 1, 'Purchase'].sum()

# Calculate gross profit
gross_profit = total_responses * profit_per_response

# Calculate total cost
total_cost = total_mailed * cost_per_mail

# Calculate net profit
net_profit = gross_profit - total_cost

# Calculate ROI
roi = net_profit / total_cost

print(f'Total Mailed: {total_mailed}')
print(f'Total Responses: {total_responses}')
print(f'Gross Profit: ${gross_profit:.2f}')
print(f'Total Cost: ${total_cost:.2f}')
print(f'Net Profit: ${net_profit:.2f}')

# Calculate ROI percentage
roi_percent = roi * 100

print(f'ROI: {roi_percent:.2f}%')

```

#### 1. Breakeven Response Rate:  
The breakeven response rate is calculated as:  

Breakeven Response Rate = Cost per Mail ÷ Profit per Response

<br>

#### 3. Profit and Cost Calculations:  
Using the targeted strategy:  
- Total Mailed: 52 customers  
- Total Responses: 15 purchases from mailed customers  

Gross Profit = Total Responses × Profit per Response  
= 15 × 20  
= $300.00

Total Cost = Total Mailed × Cost per Mail  
= 52 × 2  
= $104.00  

Net Profit = Gross Profit − Total Cost  
= 300 − 104  
= $196.00  

ROI (Return on Investment) = Net Profit ÷ Total Cost  
= 196 ÷ 104  
= 1.8846 or 188.46%
